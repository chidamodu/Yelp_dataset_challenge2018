```python
#the idea is to get the clusters for every city. So for every city name in the list I have to get the algorithm to find clusters
#and also have to link the clusters to appropriate business_id and city
def get_group(g, l):
    for key in l:
        if key in g.groups: return g.get_group(key)#when return appears the program execution stops immediately
    return pd.DataFrame()
```
